```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set to: true - to read results for fibre stiffness model run
#         false - to read results for truss model run
fibre_stiffness = True

np.set_printoptions(threshold=np.inf)

#### Read csv file
print('Reading csv files')
filepath = 'C:\\SEAK Lab\\SEAK Lab Github\\KD3M3\\Truss_AOS\\result\\AOS MOEA Runs\\'

if fibre_stiffness:
    filename_cred = 'Fibre Stiffness code run results\\Feas and Stab False\\constraint_adaptive0_credit.csv'
else: 
    filename_cred = 'Truss code run results\\Feas and Stab False\\constraint_adaptive0_credit.csv' 

full_filepath_cred = filepath + filename_cred

with open(full_filepath_cred,newline='') as csvfile:
    data = [row for row in csv.reader(csvfile)]
    iter_num_operator1_cred = list(map(float,data[0][1:]))
    credit_operator1 = list(map(float,data[1][1:]))
    iter_num_operator2_cred = list(map(float,data[2][1:]))
    credit_operator2 = list(map(float,data[3][1:]))
    iter_num_operator3_cred = list(map(float,data[4][1:]))
    credit_operator3 = list(map(float,data[5][1:]))

if fibre_stiffness:
    filename_qual = 'Fibre Stiffness code run results\\Feas and Stab False\\constraint_adaptive0_qual.csv'
else: 
    filename_qual = 'Truss code run results\\Feas and Stab False\\constraint_adaptive0_qual.csv' 

full_filepath_qual = filepath + filename_qual

with open(full_filepath_qual,newline='') as csvfile:
    data = [row for row in csv.reader(csvfile)]
    iter_num_operator1_qual = list(map(float,data[0][1:]))
    quality_operator1 = list(map(float,data[1][1:]))
    iter_num_operator2_qual = list(map(float,data[2][1:]))
    quality_operator2 = list(map(float,data[3][1:]))
    iter_num_operator3_qual = list(map(float,data[4][1:]))
    quality_operator3 = list(map(float,data[5][1:]))
    
```

```python
#operators_list = ['Remove_intersection','One_point_crossover_and_bitflip_mutation','Add_member']

#### Plot credits and quality for each operator
print('Plotting')
fig1 = plt.figure()
plt.plot(iter_num_operator1_qual,quality_operator1,'-b',label='Quality')
plt.plot(iter_num_operator1_cred,credit_operator1,'k*',label='Credit')
plt.xlabel('numbet of function evaluations')
plt.ylabel('operator quality and credit scores')
plt.title('Remove Intersection')
plt.legend(loc='upper right')
plt.show()
fig1.savefig('Remove_intersection_creds_and_quals.png')

fig2 = plt.figure()
plt.plot(iter_num_operator2_qual,quality_operator2,'-r',label='Quality')
plt.plot(iter_num_operator2_cred,credit_operator2,'k*',label='Credit')
plt.xlabel('numbet of function evaluations')
plt.ylabel('operator quality and credit scores')
plt.title('Crossover + Mutation')
plt.legend(loc='upper right')
plt.show()
fig2.savefig('Crossover_mutation_creds_and_quals.png')

fig3 = plt.figure()
plt.plot(iter_num_operator3_qual,quality_operator3,'-g',label='Quality')
plt.plot(iter_num_operator3_cred,credit_operator3,'k*',label='Credit')
plt.xlabel('numbet of function evaluations')
plt.ylabel('operator quality and credit scores')
plt.title('Add Member')
plt.legend(loc='upper right')
plt.show()
fig3.savefig('Add_member_creds_and_quals.png')
```
