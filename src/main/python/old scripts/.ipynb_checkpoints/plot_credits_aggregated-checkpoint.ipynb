```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set to: true - to read results for fibre stiffness model run
#         false - to read results for truss model run
fibre_stiffness = True

#operators_list = ['Remove_intersection','One_point_crossover_and_bitflip_mutation','Add_member']

#### Defining the functions
def read_csv_file(fib_stiff,run_num):
    filepath = 'C:\\SEAK Lab\\SEAK Lab Github\\KD3M3\\Truss_AOS\\result\\AOS MOEA Runs\\'

    if fib_stiff:
        filename = 'Fibre Stiffness code run results\\Feas and Stab False\\constraint_adaptive' + str(run_num) + '_credit.csv'
    else: 
        filename = 'Truss code run results\\Feas and Stab False\\constraint_adaptive' + str(run_num) + '_credit.csv' 

    full_filepath = filepath + filename

    with open(full_filepath,newline='') as csvfile:
        data = [row for row in csv.reader(csvfile)]
        if (data[1][0] == '.RemoveIntersection'):
            iter_num_operator1 = list(map(float,data[0][1:]))
            credit_operator1 = list(map(float,data[1][1:]))
            if (data[3][0] == 'OnePointCrossover+BitFlip'):
                iter_num_operator2 = list(map(float,data[2][1:]))
                credit_operator2 = list(map(float,data[3][1:]))
                iter_num_operator3 = list(map(float,data[4][1:]))
                credit_operator3 = list(map(float,data[5][1:]))
            elif (data[3][0] == '.AddMember'):
                iter_num_operator2 = list(map(float,data[4][1:]))
                credit_operator2 = list(map(float,data[5][1:]))
                iter_num_operator3 = list(map(float,data[2][1:]))
                credit_operator3 = list(map(float,data[3][1:]))
        elif (data[1][0] == '.AddMember'):
            iter_num_operator3 = list(map(float,data[0][1:]))
            credit_operator3 = list(map(float,data[1][1:]))
            if (data[3][0] == '.RemoveIntersection'):
                iter_num_operator1 = list(map(float,data[2][1:]))
                credit_operator1 = list(map(float,data[3][1:]))
                iter_num_operator2 = list(map(float,data[4][1:]))
                credit_operator2 = list(map(float,data[5][1:]))
            elif (data[3][0] == 'OnePointCrossover+BitFlip'):
                iter_num_operator2 = list(map(float,data[2][1:]))
                credit_operator2 = list(map(float,data[3][1:]))
                iter_num_operator1 = list(map(float,data[4][1:]))
                credit_operator1 = list(map(float,data[5][1:]))
        elif (data[1][0] == 'OnePointCrossover+BitFlip'):
            iter_num_operator2 = list(map(float,data[0][1:]))
            credit_operator2 = list(map(float,data[1][1:]))
            if (data[3][0] == '.RemoveIntersection'):
                iter_num_operator1 = list(map(float,data[2][1:]))
                credit_operator1 = list(map(float,data[3][1:]))
                iter_num_operator3 = list(map(float,data[4][1:]))
                credit_operator3 = list(map(float,data[5][1:]))
            elif (data[3][0] == '.AddMember'):
                iter_num_operator3 = list(map(float,data[2][1:]))
                credit_operator3 = list(map(float,data[3][1:]))
                iter_num_operator1 = list(map(float,data[4][1:]))
                credit_operator1 = list(map(float,data[5][1:]))
        
    return iter_num_operator1, credit_operator1, iter_num_operator2, credit_operator2, iter_num_operator3, credit_operator3

def create_full_credit_arrays(num_iter_op1, credit_op1, num_iter_op2, credit_op2, num_iter_op3, credit_op3):
    credits_full_operator1 = np.zeros(9900)
    credits_full_operator2 = np.zeros(9900)
    credits_full_operator3 = np.zeros(9900)
    index_count = 0
    for i in range(100,10000):
        if i in num_iter_op1:
            credits_full_operator1[index_count] = credit_op1[num_iter_op1.index(i)]
        elif i in num_iter_op2:
            credits_full_operator2[index_count] = credit_op2[num_iter_op2.index(i)]
        elif i in num_iter_op3:
            credits_full_operator3[index_count] = credit_op3[num_iter_op3.index(i)]
        index_count += 1
    
    iter_num_array = np.arange(100,10000,1)
    
    return iter_num_array, credits_full_operator1, credits_full_operator2, credits_full_operator3 
```

```python
#### Program operation
num_runs = 30
credits_op1_dict = {new_list: [] for new_list in range(num_runs)}
credits_op2_dict = {new_list: [] for new_list in range(num_runs)}
credits_op3_dict = {new_list: [] for new_list in range(num_runs)}

for i in range(num_runs):
    print('Reading and creating full credit arrays for run ' + str(i))
    n_iter_op1, cred_op1, n_iter_op2, cred_op2, n_iter_op3, cred_op3 = read_csv_file(fibre_stiffness,i)
    n_iter_array, cred_full_op1, cred_full_op2, cred_full_op3 = create_full_credit_arrays(n_iter_op1, cred_op1, n_iter_op2, cred_op2, n_iter_op3, cred_op3)
    credits_op1_dict[i] = cred_full_op1
    credits_op2_dict[i] = cred_full_op2
    credits_op3_dict[i] = cred_full_op3
    
## Compute the averaged credits for each operator over the runs
n_datapoints = len(credits_op1_dict[0])
credits_opt1_vals = np.zeros(num_runs)  
credits_opt2_vals = np.zeros(num_runs)
credits_opt3_vals = np.zeros(num_runs)

credits_op1_mean = np.zeros(n_datapoints)
credits_op2_mean = np.zeros(n_datapoints)
credits_op3_mean = np.zeros(n_datapoints)
credits_op1_std = np.zeros(n_datapoints)
credits_op2_std = np.zeros(n_datapoints)
credits_op3_std = np.zeros(n_datapoints)

print('Computing averaged credit values for each operator')
for i in range(n_datapoints):
    for j in range(num_runs):
        credits_opt1_current = credits_op1_dict[j]
        credits_opt2_current = credits_op2_dict[j]
        credits_opt3_current = credits_op3_dict[j]
        credits_opt1_vals[j] = credits_opt1_current[i]
        credits_opt2_vals[j] = credits_opt2_current[i]
        credits_opt3_vals[j] = credits_opt3_current[i]
    credits_op1_mean[i] = np.mean(credits_opt1_vals)
    credits_op2_mean[i] = np.mean(credits_opt2_vals)
    credits_op3_mean[i] = np.mean(credits_opt3_vals)
    credits_op1_std[i] = np.std(credits_opt1_vals)
    credits_op2_std[i] = np.std(credits_opt2_vals)
    credits_op3_std[i] = np.std(credits_opt3_vals)
    
    
## Plotting
print('Plotting')    
plt.figure()
plt.plot(n_iter_array, credits_op1_mean, 'b-', label='Remove Intersection')
plt.plot(n_iter_array, credits_op1_mean + 2*credits_op1_std, 'b.')
plt.plot(n_iter_array, credits_op1_mean - 2*credits_op1_std, 'b.')
plt.plot(n_iter_array, credits_op2_mean, 'r-', label='Crossover + Mutation')
plt.plot(n_iter_array, credits_op2_mean + 2*credits_op2_std, 'r.')
plt.plot(n_iter_array, credits_op2_mean - 2*credits_op2_std, 'r.')
plt.plot(n_iter_array, credits_op3_mean, 'g-', label='Add Member')
plt.plot(n_iter_array, credits_op3_mean + 2*credits_op3_std, 'g.')
plt.plot(n_iter_array, credits_op3_mean - 2*credits_op3_std, 'g.')
plt.xlabel('numbet of function evaluations')
plt.ylabel('average operator credits over ' + str(num_runs) + ' runs')
plt.legend(loc='upper right')
plt.show()
```
